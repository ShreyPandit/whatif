```
import pandas as pd
df = pd.read_csv("match_info_final_updated_normalized.csv?token=AFQLSH4HULON2XF56IL2HYS63U6PO")
```

```
df.head()
```

```
full_list = []
for index, row in df.iterrows():
    embed_whole = []
    embed_whole.append(row.Goals)
    embed_whole.append(row["On Target"])
    embed_whole.append(row["Off Target"])
    embed_whole.append(row["Blocks"])
    embed_whole.append(row["Corners"])
    embed_whole.append(row["Offside"])
    embed_whole.append(row["Possession"])
    embed_whole.append(row["Passes"])
    embed_whole.append(row["Passing Accuracy"])
    embed_whole.append(row["Distance Covered"])
    embed_whole.append(row["Tackles"])
    embed_whole.append(row["Clearances"])
    embed_whole.append(row["Fouls"])
    embed_whole.append(row["Yellow Cards"] + row["Red Cards"])
    full_list.append(embed_whole)
```

```
df["Full Embeddings"] = full_list
```

```
team_dict = {}
for i,row in df.iterrows():
    if row["Teams"] not in team_dict:
        team_dict[row["Teams"]] = []
        team_dict[row["Teams"]].append(row["Full Embeddings"])
    else:
        team_dict[row["Teams"]].append(row["Full Embeddings"])
```

```
import numpy as np
team_name = []
team_embed = []
for i in team_dict:
    temp = np.mean(np.array(team_dict[i]), axis=0)
    team_name.append(i)
    team_embed.append(temp)
    
```

```
fd = pd.DataFrame()
fd["Teams"] = team_name
fd["Embeddings"] = team_embed
```

```
fd.to_csv("Team_embeddings.csv",index=False)
```
