```python
import pandas as pd
import numpy as np
```

```python
matches_df = pd.read_csv('./../Data/matches_info.csv')
players_df = pd.read_csv('./../Data/players_info.csv')
```

```python
matches_df.head()
```

```python
matches_df['Total Attempts'] = matches_df['On Target'] + matches_df['Off Target']
```

```python
matches_df['Passing Accuracy'] = matches_df['Passes Completed'] / matches_df['Passes']
```

```python
matches_df.head()
```

```python
matches_df.info()
```

```python
matches_df[matches_df['Possession'] != 'Not Provided'].shape
```

```python
matches_df = matches_df.replace(r'Not Provided', np.nan, regex=True)
```

```python
matches_df['Teams'] = matches_df["Teams"].astype('category')
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
matches_df['Temp'] = lb.fit_transform(matches_df['Teams'])
matches_df.head()
```

```python
matches_df['Woodwork'] = matches_df['Woodwork'].fillna(matches_df.groupby('Temp')['Woodwork'].transform('mean'))
```
