```python
teams_dict_flag = {}
teams_dict_idx = {}
teams_dict_data = {}

for i in range(matches_df.shape[0]):
    team_name = matches_df.iloc[i,0]
    teams_dict_idx[team_name] = []
    teams_dict_flag[team_name] = 0
    
for i in range(matches_df.shape[0]):
    team_name = matches_df.iloc[i,0]
    if teams_dict_flag[team_name] == 1:
        temp = []
        for j in range(i):
            if matches_df.iloc[j,0] == team_name:
                temp.append(matches_df.iloc[j,1:-1].values)
        list_avg = np.sum(temp,axis = 0) / len(temp)
        if team_name == "Paris":
            print(list_avg)
            print("added here")
        list_avg2 = np.append(list_avg,matches_df.iloc[i,-1])
        teams_dict_data[i] = list_avg2
        
        
    else:
        teams_dict_data[i] = matches_df.iloc[i,1:].values
        if team_name == "Paris":
            print("lol here")
    teams_dict_idx[team_name].append(i)
    teams_dict_flag[team_name] = 1
    
```
